```python
train = pd.read_csv("/content/train.csv")
train.head()
```

```python
d_cols = ['Name', 'Ticket','Cabin']
train = train.drop(d_cols, axis = 1)
train.info()
```

```python
train = train.dropna()
```

```python
train.info()
```

```python
dummy = []
columns = ['Pclass', 'Sex', 'Embarked']
for col in columns:
   dummy.append(pd.get_dummies(train[col]))
```

```python
titanic_dummies = pd.concat(dummy, axis=1)
```

```python
train = pd.concat((train,titanic_dummies), axis=1)
```

```python
train = train.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
```

```python
train.info()
```

```python
train['Age'] = train['Age'].interpolate()
```

```python
train.info()
```

```python
X = train.values
y = train['Survived'].values
X = np.delete(X, 1, axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```
